## Prediction of the flowers

```python
from tensorflow.keras.models import load_model
```

```python
from tensorflow.keras.preprocessing import image
```

```python
model = load_model("flowerss.h5")
```

```python
img = image.load_img("sunflowers.jpg",target_size=(256,256))
```

```python
img
```

```python
type(img)
```

```python
x = image.img_to_array(img)
```

```python
x
```

```python
x.shape
```

```python
import numpy as np
```

```python
# convolution expects 4D
x = np.expand_dims(x,axis=0)
```

```python
x.shape
```

```python
pred_prob = model.predict(x)
```

```python
pred_prob
```

```python
class_name=["daisy","dandelion","rose","sunflower","tulip"]
pred_id = pred_prob.argmax(axis=1)[0]
```

```python
pred_id
```

```python
print("The predicted flower is",str(class_name[pred_id]))
```
